```python
# remove rows by filtering
dfx =df[(df['Area'] != 'Cork') & (df['Area'] != 'Dublin') & (df['Area'] != 'Galway') & (df['Area'] != 'Limerick')
        & (df['Area'] != 'Waterford') & (df['Area'] != 'Other areas') & (df['Statistic'] != 'Second Hand House Prices')]
dfx[df['Quarter'].isin(['2015Q1', '2015Q2', '2015Q3', '2015Q4', '2016Q1', '2016Q2', '2016Q3', '2016Q4'])]
#dfx = dfx.copy(deep=True)
dfx
```

```python
dfx1 = dfx.copy(deep=True)
a = ['2015Q1', '2015Q2', '2015Q3', '2015Q4', '2016Q1', '2016Q2', '2016Q3', '2016Q4']
dfx1 = dfx1[dfx1['Quarter'].isin(a)]
dfx1
```

```python
df1.drop(['code', 'houses_sold', 'no_of_crimes', 'borough_flag'], axis=1, inplace=True)
```

```python
df1
```

```python
#rename VALUE to indicate currency of transaction
df1 = df1.rename(columns={"average_price":"Price"})
#dropping rows with missing values
df1 = df1.dropna(axis=0)
df1
```

```python
#select rows where 'area' column is equal to 'england'
df1 = df1.loc[df1['area'] == 'england']
df1
```

```python
df1 = df1[(df1['date'] > '2014-12-01') & (df1['date'] < '2017-01-01')]
pd.to_datetime(df1['date'])
df1['Quarter'] = pd.PeriodIndex(df1.date, freq='Q')
df1
```

```python
# Use GroupBy() to compute the sum
df1 = df1.groupby('Quarter').mean()
df1
```

```python
df1.reset_index(inplace=True)
df1 = df1.rename(columns = {'index':'Quarter'})
df1 = df1.copy(deep=True)
df1
```

```python
frame = [dfx1, df1]
result = pd.concat(frame)
result.reset_index(inplace=True)
result
```

```python
result.Area = result.Area.fillna('england')
result.Price = result.Price.apply(np.ceil) 
result.drop(['index', 'Statistic'], axis=1, inplace=True)

result        
```

```python
sns.barplot(data=result, x='Price', y='Quarter', hue='Area')
```

```python
sns.lineplot(data=result, x='Area', y='Price')
```

```python
sns.lmplot(data=result, x='Price', y='Area', x_estimator=np.mean, ci=90);
```
